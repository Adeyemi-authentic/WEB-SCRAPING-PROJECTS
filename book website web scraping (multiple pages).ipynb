```python
from bs4 import BeautifulSoup

import requests
```

```python
url="https://books.toscrape.com/catalogue/category/books_1/page-2.html"

page=requests.get(url)

soup=BeautifulSoup(page.content,'html.parser')
```

```python
print(soup)
```

```python
order_list=soup.find('ol')
```

```python
order_list
```

```python
articles=order_list.find_all('article',class_='product_pod')
```

```python
books=[]
for i in range(2,6):
    url="https://books.toscrape.com/catalogue/category/books_1/page-i.html"

    page=requests.get(url)

    soup=BeautifulSoup(page.content,'html.parser')
    order_list=soup.find('ol')
```

```python




for article in articles:
    image=article.find('img')
    title=image.attrs['alt']
    #print(title)
    
    star=article.find('p')
    star=star['class'][1]
    #print(star)
    price=article.find('p',class_='price_color').text
    #print(price)
  
    price=float(price[1:])
    
    books.append([title,price,star])
    
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(books,columns=['title','price','star ratings'])
df
```
